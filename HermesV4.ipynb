```python
import re, json, random
from collections import Counter

def clean(text):
    if not text:
        return None
    text = str(text).strip()
    text = re.sub(r'\s+', ' ', text)
    return text if 10 < len(text) < 400 else None

def extract_pairs_from_dialogue(dialogue, scammer_tag, victim_tag, source):
    """Extract scammer→victim consecutive pairs from a dialogue string."""
    pairs = []
    if not dialogue:
        return pairs

    # Split into turns using all known speaker tags
    # Replace tags with a delimiter we can split on
    text = str(dialogue)
    for tag in [scammer_tag + ":", victim_tag + ":"]:
        text = text.replace(tag, f"\n{tag}")

    lines = [l.strip() for l in text.split("\n") if l.strip()]
    turns = []
    for line in lines:
        if line.startswith(scammer_tag + ":"):
            turns.append(("scammer", line.replace(scammer_tag + ":", "").strip()))
        elif line.startswith(victim_tag + ":"):
            turns.append(("victim", line.replace(victim_tag + ":", "").strip()))

    # Extract consecutive scammer → victim pairs
    for i in range(len(turns) - 1):
        if turns[i][0] == "scammer" and turns[i+1][0] == "victim":
            scammer_msg = clean(turns[i][1])
            victim_reply = clean(turns[i+1][1])
            if scammer_msg and victim_reply:
                pairs.append({
                    "input": scammer_msg,
                    "output": victim_reply,
                    "source": source
                })
    return pairs

all_pairs = []

# ── DS1: Innocent/Suspect format, only scam rows (labels==1) ────────────────
print("Processing DS1...")
for row in ds1:
    # labels==1 means scam, labels==0 means non-scam
    # We want BOTH — scam for training Hermes, non-scam can be filtered
    if str(row.get("labels", "0")) == "1" or True:  # use all for now
        pairs = extract_pairs_from_dialogue(
            row.get("dialogue", ""),
            scammer_tag="Suspect",
            victim_tag="Innocent",
            source="ds1_multi_agent"
        )
        all_pairs.extend(pairs)
print(f"  DS1 pairs: {len(all_pairs)}")
after_ds1 = len(all_pairs)

# ── DS2: Same Innocent/Suspect format ───────────────────────────────────────
print("Processing DS2...")
for row in ds2:
    pairs = extract_pairs_from_dialogue(
        row.get("dialogue", ""),
        scammer_tag="Suspect",
        victim_tag="Innocent",
        source="ds2_single_agent"
    )
    all_pairs.extend(pairs)
print(f"  DS2 pairs: {len(all_pairs) - after_ds1}")
after_ds2 = len(all_pairs)

# ── DS3: Person A/Person B format, label==1 means scam ─────────────────────
print("Processing DS3...")
for row in ds3:
    if row.get("label") == 1:  # scam conversations only
        pairs = extract_pairs_from_dialogue(
            row.get("conversation", ""),
            scammer_tag="Person A",
            victim_tag="Person B",
            source="ds3_scam_baiter"
        )
        all_pairs.extend(pairs)
        # Also try reverse (sometimes victim speaks first)
        pairs_rev = extract_pairs_from_dialogue(
            row.get("conversation", ""),
            scammer_tag="Person B",
            victim_tag="Person A",
            source="ds3_scam_baiter_rev"
        )
        all_pairs.extend(pairs_rev)
print(f"  DS3 pairs: {len(all_pairs) - after_ds2}")
after_ds3 = len(all_pairs)

# ── DS4: SMS spam — use quick reply templates ────────────────────────────────
print("Processing DS4...")
QUICK_REPLIES = [
    "OTP? Which app should I open beta? There are so many on my phone.",
    "My son told me never share anything on phone. Are you sure this is safe?",
    "Wait wait, the app is loading. My BSNL connection is very slow today.",
    "I need to write this down. Where is my pen... Seema! Pen kahan hai!",
    "KYC? I did this last year only. Does it expire so fast beta?",
    "I have two accounts, one SBI and one post office. Which one do you need?",
    "My spectacles are not with me. Can you repeat slowly please beta?",
    "My grandson handles all these digital things. He just went to school.",
    "Please don't block. I am senior citizen, pension comes in this account.",
    "Let me call my son Suresh first. He handles all these important matters.",
    "I clicked but it is showing some error. Should I click again beta?",
    "Which browser should I use? I have Chrome and also another blue one.",
    "The page is asking me to download something. Is that safe beta?",
    "I forgot my PIN beta. Can I reset it? Will that take very long?",
    "Sorry can you repeat? There is noise from my neighbour's TV.",
    "I did not understand properly. My hearing is not so good these days.",
    "Haan I am listening. My phone battery was low, just put on charging.",
    "Can you speak a little slowly? I am an old man, things take time.",
    "Beta ek minute, meri chai aa gayi. Haan ab batao kya karna hai.",
    "I think I pressed wrong button. Can we please start from beginning?",
]

reply_idx = 0
for row in ds4:
    if row.get("label") == 1 or str(row.get("label","")).lower() == "spam":
        text = clean(row.get("sms", row.get("text", "")))
        if text:
            reply = QUICK_REPLIES[reply_idx % len(QUICK_REPLIES)]
            reply_idx += 1
            all_pairs.append({
                "input": text,
                "output": reply,
                "source": "sms_spam"
            })
print(f"  DS4 pairs: {len(all_pairs) - after_ds3}")

print(f"\nTotal pairs: {len(all_pairs)}")
print("\nSource breakdown:")
dist = Counter(p["source"] for p in all_pairs)
for src, count in sorted(dist.items(), key=lambda x: -x[1]):
    print(f"  {src:30s}: {count}")

# Quality check
print("\n=== Sample pairs from each source ===")
seen_sources = set()
for p in random.sample(all_pairs, min(20, len(all_pairs))):
    if p["source"] not in seen_sources:
        seen_sources.add(p["source"])
        print(f"\nSource: {p['source']}")
        print(f"  Scammer: {p['input'][:80]}")
        print(f"  Reply:   {p['output'][:80]}")
```

```python
import re
from collections import Counter

def is_good_pair(pair):
    inp = pair["input"].strip()
    out = pair["output"].strip()

    # Basic length checks
    if len(inp) < 10 or len(out) < 15:
        return False
    if len(inp) > 350 or len(out) > 350:
        return False

    # Output is just repeating input exactly
    if out.lower().strip() == inp.lower().strip():
        return False

    # Must have at least some alphabetic content
    if not any(c.isalpha() for c in inp):
        return False
    if not any(c.isalpha() for c in out):
        return False

    # Filter outputs that are clearly scammer-like
    # (DS3 sometimes has scammer lines as both input AND output)
    scammer_phrases = [
        "time is of the essence",
        "we must act swiftly",
        "i cannot disclose",
        "due to company policies",
        "we need to verify your",
        "billing discrepancies",
        "act immediately",
        "your account has been compromised",
        "we are from the bank",
        "press 1 to speak",
    ]
    out_lower = out.lower()
    if any(phrase in out_lower for phrase in scammer_phrases):
        return False

    return True

clean_pairs = [p for p in all_pairs if is_good_pair(p)]
print(f"Before cleaning: {len(all_pairs)}")
print(f"After cleaning:  {len(clean_pairs)}")

# Deduplicate by input text
seen_inputs = set()
deduped = []
for p in clean_pairs:
    key = p["input"].lower().strip()[:80]
    if key not in seen_inputs:
        seen_inputs.add(key)
        deduped.append(p)

print(f"After dedup:     {len(deduped)}")

# Show source breakdown after cleaning
print("\nSource breakdown after cleaning:")
dist = Counter(p["source"] for p in deduped)
for src, count in sorted(dist.items(), key=lambda x: -x[1]):
    print(f"  {src:30s}: {count}")

# Quality check — sample from each source
print("\n=== Quality check — samples per source ===")
seen_sources = set()
for p in deduped:
    if p["source"] not in seen_sources:
        seen_sources.add(p["source"])
        print(f"\nSource: {p['source']}")
        print(f"  Scammer: {p['input'][:90]}")
        print(f"  Reply:   {p['output'][:90]}")
    if len(seen_sources) >= 5:
        break
```

```python
from datasets import Dataset

dataset = Dataset.from_list([
    {
        "input_text":  f"respond as confused elderly indian man to scammer: {p['input']}",
        "target_text": p["output"]
    }
    for p in final_pairs
])

print(f"Dataset: {len(dataset)} examples")
print("\n3 samples:")
for i in [0, 100, 500]:
    print(f"\n  Input:  {dataset[i]['input_text'][:90]}")
    print(f"  Target: {dataset[i]['target_text'][:90]}")
```

```python
from transformers import T5Tokenizer, T5ForConditionalGeneration
from peft import get_peft_model, LoraConfig, TaskType
import torch, gc

gc.collect()
torch.cuda.empty_cache()

MODEL_NAME = "t5-base"   # Better than t5-small, still only 250MB

print("Loading T5-base...")
tokenizer = T5Tokenizer.from_pretrained(MODEL_NAME)

model = T5ForConditionalGeneration.from_pretrained(
    MODEL_NAME,
    torch_dtype=torch.float16,
    device_map="auto"
)

lora_config = LoraConfig(
    task_type=TaskType.SEQ_2_SEQ_LM,
    r=32,
    lora_alpha=64,
    lora_dropout=0.05,
    target_modules=["q", "v", "k", "o"],   # Train more attention layers
    bias="none"
)

model = get_peft_model(model, lora_config)
model.print_trainable_parameters()
print("T5-base with LoRA ready!")
```

```python
MAX_INPUT  = 128
MAX_TARGET = 80

def tokenize(batch):
    # Tokenize inputs
    model_inputs = tokenizer(
        batch["input_text"],
        max_length=MAX_INPUT,
        truncation=True,
        padding="max_length"
    )

    # Tokenize targets (labels)
    labels = tokenizer(
        text_target=batch["target_text"],   # ✅ NEW WAY
        max_length=MAX_TARGET,
        truncation=True,
        padding="max_length"
    )

    # Replace padding token id with -100 (important)
    label_ids = labels["input_ids"]
    label_ids = [
        [-100 if token == tokenizer.pad_token_id else token for token in label]
        for label in label_ids
    ]

    model_inputs["labels"] = label_ids
    return model_inputs


tokenized = dataset.map(
    tokenize,
    batched=True,
    remove_columns=["input_text", "target_text"],
    desc="Tokenizing"
)

print("Tokenized!", tokenized)
```

```python
from transformers import (
    TrainingArguments, Trainer,
    DataCollatorForSeq2Seq,
    TrainerCallback, EarlyStoppingCallback
)

gc.collect()
torch.cuda.empty_cache()

CHECKPOINT_DIR = "/content/drive/MyDrive/HermesV4/checkpoints"

training_args = TrainingArguments(
    output_dir=CHECKPOINT_DIR,
    num_train_epochs=10,
    per_device_train_batch_size=16,
    gradient_accumulation_steps=2,
    learning_rate=2e-4,   # ✅ FIXED
    warmup_ratio=0.05,
    weight_decay=0.01,
    eval_strategy="steps",
    eval_steps=200,
    save_strategy="steps",
    save_steps=200,
    save_total_limit=5,
    load_best_model_at_end=True,
    metric_for_best_model="eval_loss",
    greater_is_better=False,
    fp16=True,
    logging_steps=50,
    logging_first_step=True,
    report_to="none",
)

split = tokenized.train_test_split(test_size=0.05, seed=42)
print(f"Train: {len(split['train'])} | Eval: {len(split['test'])}")

class HermesCallback(TrainerCallback):
    def on_save(self, args, state, control, **kwargs):
        print(f"  Checkpoint saved → step {state.global_step} → Drive")
        return control
    def on_log(self, args, state, control, logs=None, **kwargs):
        if not logs:
            return control
        loss      = logs.get("loss", None)
        eval_loss = logs.get("eval_loss", None)
        step      = state.global_step
        epoch     = round(state.epoch or 0, 2)
        if loss:
            msg = f"  Step {step:>5} | Epoch {epoch:>5} | Train: {loss:.4f}"
            if eval_loss:
                msg += f" | Eval: {eval_loss:.4f}"
            print(msg)
        return control

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=split["train"],
    eval_dataset=split["test"],
    data_collator=DataCollatorForSeq2Seq(
        tokenizer,
        model=model,
        padding=True,
        label_pad_token_id=-100
    ),
    callbacks=[
        HermesCallback(),
        EarlyStoppingCallback(early_stopping_patience=5)
    ]
)

print("\n" + "="*55)
print("  HERMES V4 — Final Training")
print(f"  Model:      T5-base")
print(f"  Dataset:    {len(split['train'])} pairs")
print(f"  Epochs:     10 (early stop if no improvement)")
print(f"  Saves:      every 200 steps → Drive")
print(f"  Target:     eval loss below 0.8")
print("="*55 + "\n")

result = trainer.train()

print("\n" + "="*55)
print("  TRAINING COMPLETE!")
print(f"  Final train loss: {result.training_loss:.4f}")
metrics = trainer.evaluate()
print(f"  Final eval loss:  {metrics['eval_loss']:.4f}")
print("="*55)

# Save immediately
print("\nSaving to Drive...")
trainer.save_model("/content/hermes_v4")
tokenizer.save_pretrained("/content/hermes_v4")
trainer.save_model("/content/drive/MyDrive/HermesV4/model")
tokenizer.save_pretrained("/content/drive/MyDrive/HermesV4/model")
print("Saved!")
```

```python
print("=== Testing Hermes V4 ===\n")
model.eval()

test_msgs = [
    "Give me your OTP now",
    "Send money to paytm@ybl immediately",
    "Your KYC has expired",
    "Account block ho jayega",
    "You have won a lottery of Rs 50000",
    "Click this link to verify your account",
    "Share your Aadhaar number",
    "This is very urgent act now",
    "You are under arrest for fraud",
    "Install this app and share your screen",
    "OTP bhejo abhi warna account band ho jayega",
    "Paisa bhejo turant Rs 10000",
]

for msg in test_msgs:

    prompt = f"""
You are a confused elderly Indian man.
You ask different questions every time.
Do not repeat the same sentence.

Scammer: {msg}
Hermes:
"""

    inputs = tokenizer(
        prompt,
        return_tensors="pt",
        max_length=128,
        truncation=True
    ).to(model.device)

    with torch.no_grad():
        outputs = model.generate(
            **inputs,   # ✅ FIXED
            max_new_tokens=60,
            do_sample=True,
            temperature=0.9,
            top_p=0.92,
            top_k=60,
            repetition_penalty=1.4,
            no_repeat_ngram_size=3,
            pad_token_id=tokenizer.eos_token_id,
        )

    reply = tokenizer.decode(outputs[0], skip_special_tokens=True)

    print(f"Scammer: {msg}")
    print(f"Hermes:  {reply}")
    print()
```
